# pylab——[改变figure的布局和大小](https://blog.csdn.net/panda1234lee/article/details/52311593)

搜索关键词：plt.xticks()/plt.yticks()设置轴记号

```python
from pylab import *

# 创建一个 8 * 6 点（point）的图，并设置分辨率为 80
fig1 = figure(figsize=(8,  6), dpi=80)
subplot(1, 1, 1)
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
plot(X, C, color="blue", linewidth=1.0, linestyle="-")
plot(X, S, color="r", lw=4.0, linestyle="-")

plt.axis([-4, 4, -1.2, 1.2])  # 设置背景的长宽，一般比有限长曲线的边界长
xticks([-np.pi/3, -np.pi / 2, 0, np.pi / 2, np.pi],  [r'$-\pi$', r'$-\pi/2$',    # plt.xticks()/plt.yticks() 指定轴的显示值
													r'$0$', r'$+\pi/2$', r'$+\pi$'])              # plt.yticks([指定位置], [指定显示名称])
yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

show()
```

# Matplotlib——[设置 fig 大小](https://blog.csdn.net/cdqn10086/article/details/72403899)

方法一：

```python
import numpy as np  
import matplotlib.pyplot as plt

fig2 = plt.figure(figsize=(20,10))

t = np.arange(0.0, 5.0, 0.01)  
s = np.cos(2*np.pi*t)  
plt.plot(t, s, lw=3, color='orange')  
plt.ylim(-2,2)  
plt.show() 
```

方法二：

```python
import numpy as np  
import matplotlib.pyplot as plt

fig3, axe = plt.subplots(figsize=(10,6))

t = np.arange(0.0, 5.0, 0.01)  
s = np.sin(2*np.pi*t)  
plt.plot(t, s, lw=4, color='green')  
plt.ylim(-2,2)  
plt.show() 
```

方法三:

Ps: gca()和gcf()(get current axes/figures)分别获取当前axes和figures的对象。

```python
import numpy as np  
import matplotlib.pyplot as plt

fig4, axe4 = plt.subplots()

t = np.arange(0.0, 5.0, 0.01)  
s = np.sin(2*np.pi*t)
c = np.cos(2*np.pi*t)
plt.plot(t, s, lw=3, color='red')  
plt.plot(t, c, lw=4, color='gray')
#plt.ylim(-2,2)

#fig4 = plt.gcf()  # 若一开始没有给定 fig4 的句柄，可以用 plt.gcf() 获取
#axe4 = plt.gca()  # 同理，可以用 plt.gcca() 获取 axe 句柄
fig4.set_size_inches(10, 7)
axe4.set_ylim(-2,2)

plt.show() 
```
